```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from google.colab import files

values = [6.53, 45.824, 65.08, 8.85]
total = 100

# Calculate remaining percentages
remaining = [total - value for value in values]

# Use Seaborn color palette for nice, dark colors
colors = sns.color_palette('deep')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

# Create bar plot
bars1 = plt.bar(np.arange(len(values)), values, color=colors[1], edgecolor='black', linewidth=1.5, label='Prefetched')
bars2 = plt.bar(np.arange(len(values)), remaining, bottom=values, color=colors[0], edgecolor='black', linewidth=1.5, label='Not Prefetched')

# Set labels and title
plt.xlabel('Benchmark', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.title('Pointer faults prefetched vs not prefetched', fontsize=14)

# Set X axis tick labels
labels = ['canneal', 'linkedlist', 'bfs', 'redis']
plt.xticks(np.arange(len(values)), labels, fontsize=14)

# Set y-axis tick font size
plt.yticks(fontsize=10)

# Add bold crossed line inside each bar
for bar in bars1:
    bar.set_hatch('/')
    bar.set_linewidth(1.5)

# Add bold crossed line inside each bar
for bar in bars2:
    bar.set_hatch('xx')
    bar.set_linewidth(1.5)

# Add legend
plt.legend()

# Save the figure as a PDF file
plt.savefig('divided_bar_graph.pdf', bbox_inches='tight')

# Generate a download link for the PDF file
files.download('divided_bar_graph.pdf')

# Show the plot
plt.show()
```
